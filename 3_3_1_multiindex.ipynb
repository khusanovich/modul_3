### `Hierarchical indexing`

```python
data = pd.Series(np.random.randn(9), index=['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd']) # odatiy indeksli Series
data.index
```

```python
# irarxik indeksli Series
data1 = pd.Series(np.random.randn(9), 
                  index=[['a', 'a', 'a', 'b', 'b', 'c', 'c', 'd', 'd'],[1, 2, 3, 1, 3, 1, 2, 2, 3]])
data1.index
```

### Dunyo shaharlarining havo va suv sifati ma'lumotlari ustida amallar bajarish

```python
!git clone https://github.com/MansurCompAI/Practicum_Datasets.git # Githubdagi reponi nusxalab oladi
```

```python
df = pd.read_csv("/content/Practicum_Datasets/world_cities_air_water_quality.csv")
df
```

```python
df.info()
```

```python
columns = ['City', 'Region', 'Country', 'AirQuality', 'WaterPollution']
df.columns = columns
df
```

```python
df['Region'] = df['Region'].map(lambda x: x.replace('"', "")) # Region ustuniga tegishli ma'lumotlardan " belgisini olib tashlash
df
```

```python
df['Country'] = df['Country'].map(lambda x: x.replace('"', "")) # Country ustuniga tegishli ma'lumotlardan " belgisini olib tashlash
df
```

```python
df.to_csv("Updated_dataset.csv", index=False)
```

```python
df[df['Country']=='Germany']
```

```python
df['Country'][2]
```

```python
df['Country'][3]
```

```python
df['Country'] = df['Country'].map(lambda x: x.strip()) # Country ustuniga tegishli ma'lumotlardan bo'sh belgi(whitespace)ni belgisini olib tashlash
df
```

```python
df[df['Country']=='Germany']
```

```python
df[df['Country']=='Uzbekistan']
```

```python
df.to_csv("Updated_dataset.csv", index=False) # faylni saqlab qo'yish
```

### `pivot_table` orqali irarxik indekslash

```python
df.index
```

```python
df1 = pd.pivot_table(df, index=['Country', 'City']) # irarxik indeksli ma'lumotlar to'plamiga o'tish
```

```python
df1
```

```python
df1.loc[['Uzbekistan', 'Germany', 'Tajikistan']] # Ba'zi davlatlarga tegishli ma'lumotlar
```

### Ustunlarni irarxik ko'rinishga keltirish

```python
df2 = df1.copy()
df2
```

```python
df2['AirPollution'] = 100-df2['AirQuality'] # AirPollution ustuniga teskari ko'rsatgichli ustun yaratish
df2
```

```python
df2['WaterQuality'] = 100-df2['WaterPollution'] # WaterQuality ustuniga teskari ko'rsatgichli ustun yaratish
df2
```

## Irarxik ustunli ma'lumotlar to'plami

```python
df3 = pd.DataFrame(df2.values, index = df2.index, columns= [['AIR', 'AIR', 'WATER', 'WATER'], ['AirQuality', 'AirPollution', 'WaterQuality', 'WaterPollution']]) # 
df3
```
